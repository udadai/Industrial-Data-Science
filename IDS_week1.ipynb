```python
import pandas as pd

# 데이터 로드
df = pd.read_csv('Iris.csv')
print(df.head())
```

```python
df.info()
```

```python
# iris 데이터 종별로 통계값 출력
df.groupby('Species').describe()
```

- SepalLength가 5.0cm 이상 5.7cm 이하
- PetalLength가 1.5cm 이하이거나 3.0cm 이상
- 종별 컬럼 통계값은 describe함수 사용

```python
df_filtered = df[(df['SepalLengthCm'] >= 5.0) & (df['SepalLengthCm'] <= 5.7) & ((df['PetalLengthCm'] <= 1.5) | (df['PetalLengthCm'] >= 3.0))]
df_filtered.groupby('Species').describe()
```

```python
df['Species'].unique()
```

```python
df_setosa = df_filtered[df_filtered['Species'] == 'Iris-setosa']
df_versicolor = df_filtered[df_filtered['Species'] == 'Iris-versicolor']
df_virginica = df_filtered[df_filtered['Species'] == 'Iris-virginica']
```

```python
# 종이 'setosa'일 때 통계값
df_setosa.describe()
```

```python
# 종이 'versicolor'일 때 통계값
df_versicolor.describe()
```

```python
# 종이 'virginica'일 때 통계값
df_virginica.describe()
```
